```python
import pandas as pd #read files
from pathlib import Path
```

```python
#merging NonProducts into single csv
# List of CSV file names to merge
file_path = '/Users/Tico/Documents/Masters/LaMona/CSV/Non-Products/'
file_names = ['missdivaChildren.csv','yoursChildren.csv','tekshopChildren.csv','skatesChildren.csv','portopetsChildren.csv','rollerbladeChildren.csv','cabinetshopChildren.csv']

# Create an empty DataFrame to store merged data
nonProducts_df = pd.DataFrame()

# Merge CSV files
for filename in file_names:
    productCSV = pd.read_csv(file_path+ filename)
    nonProducts_df = pd.concat([nonProducts_df, productCSV], ignore_index=True)

nonProducts_df['category'] = "Not Product"
# Save the merged DataFrame to a new CSV file
nonProducts_df.to_csv('/Users/Tico/Documents/Masters/LaMona/ModelTraining/NonProducts.csv', index=False)
```

```python
#merging NonProducts into single csv
# List of CSV file names to merge
file_path = '/Users/Tico/Documents/Masters/LaMona/CSV/Products/'
file_names = ['missdivaHeels.csv','yoursCrops&Shorts.csv','tekshopLaptop.csv','skatesScooters.csv','portopetsCats.csv','rollerbladeSkates.csv','cabinetshopItems.csv']

# Create an empty DataFrame to store merged data
products_df = pd.DataFrame()

# Merge CSV files
for filename in file_names:
    productCSV = pd.read_csv(file_path+ filename)
    products_df = pd.concat([products_df, productCSV], ignore_index=True)
products_df['category'] = "Product"
# Save the merged DataFrame to a new CSV file
products_df.to_csv('/Users/Tico/Documents/Masters/LaMona/ModelTraining/Products.csv', index=False)
```
